# Training Material: Creating RGB Color Composites from ALOS-2 Data for Change Detection

## I. Introduction

This training material demonstrates how to create RGB color composites from ALOS-2 SAR data for change detection analysis using Python. RGB composites assign different image bands or data from different acquisition dates to red, green, and blue channels, allowing for visualization of changes over time. This technique is particularly useful for monitoring urban development, deforestation, disaster assessment, and other applications where temporal change analysis is crucial.

```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt
```

## I. Loading and Reading ALOS-2 SAR data.

```python
# Paths to the images for the RGB channels
# Set these to your specific files or modify to load bands from a single multi-band file
red_band_path = "/home/jovyan/shared/Arissara/genAI/processed_images/processed_IMG-HH-ALOS2290760720-191011-UBSR2.1GUA.tif"
green_band_path = "/home/jovyan/shared/Arissara/genAI/processed_images/processed_IMG-HH-ALOS2294160720-191103-UBSR2.1GUA.tif"
blue_band_path = "/home/jovyan/shared/Arissara/genAI/processed_images/processed_IMG-HH-ALOS2294160720-191103-UBSR2.1GUA.tif"

# Open each band and read the data
with rasterio.open(red_band_path) as red_src:
    red = red_src.read(1)  # Read the red band

with rasterio.open(green_band_path) as green_src:
    green = green_src.read(1)  # Read the green band

with rasterio.open(blue_band_path) as blue_src:
    blue = blue_src.read(1)  # Read the blue band
```

## II. Creating RGB Composites for Change Detection.

We will create a temporal composite by assigning different dates to different color channels.

```python
# Stack the bands into an RGB image array
rgb_image = np.stack([red, green, blue], axis=-1)

# Normalize the bands for better visualization (optional)
rgb_image = (rgb_image - rgb_image.min()) / (rgb_image.max() - rgb_image.min())  # Normalize to [0, 1]
```

## III. Visualization.

```python
# Display the RGB composite
plt.figure(figsize=(10, 10))
plt.imshow(rgb_image)
plt.title("RGB Color Composite")
plt.axis("off")
plt.show()
```
